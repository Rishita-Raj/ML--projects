```python

import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
delhi_dataset = pd.read_csv('/content/delhi_power_demand_dataset.csv')
```

```python
delhi_dataset.head()
```

```python
delhi_dataset.tail()
```

```python
delhi_dataset.shape
```

```python
delhi_dataset.describe()
```

```python

# number of missing value in each column
delhi_dataset.isnull().sum()
```

```python
# Check and convert 'is_weekend' column to string if needed
delhi_dataset['is_weekend'] = delhi_dataset['is_weekend'].astype(str)
```

```python
delhi_dataset['is_holiday'] = delhi_dataset['is_holiday'].astype(str)
```

```python
# Replace 'False' with 0 and 'True' with 1
delhi_dataset.replace({"is_weekend": {'False': 0, 'True': 1},"is_holiday": {'False':0, 'True': 1}}, inplace=True)
```

```python
# Verify the replacement
print(delhi_dataset.head())
```

```python
# separating the data and label
X = delhi_dataset.drop(columns=['Unnamed: 0','is_weekend'],axis=1)
Y = delhi_dataset['is_weekend']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
#training the support vector machine model
classifier.fit(X_train,Y_train)
```

```python
# accuracy score on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data : ', training_data_accuracy)
```

```python
# accuracy score on test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on test data : ', test_data_accuracy)
```
